## Data  Insights 

### a. Common job titles in data science jobs

```python
df['job_title_sim'].value_counts().nlargest(5)
```

The most popular 5 job titles are:

1- data scientist                                                       
2- other scientist                                                         
3- data engineer                                               
4- analyst                                                          
5- machine learning engineer

```python
plt.figure(figsize=(7,7))
df['job_title_sim'].value_counts().nlargest(5).plot(kind='bar', color = '#8B7355', edgecolor='black')
plt.title("Job Title",fontsize=15)
```

Data Science related jobs includes titles: Data engineer, analyst and machine learning engineer

### b. Avarage salary in different states

```python
avg_salary = df.groupby("Job Location")["Avg Salary(K)"].mean().sort_values(ascending=False)[0:10]
avg_salary = avg_salary.reset_index().rename(columns={"Job Location":"Job Location","Avg Salary(K)":"Avg Salary($)"})
```

```python
fig, ax = plt.subplots(figsize=(7,7))
sns.barplot(x="Job Location", y = "Avg Salary($)",data=avg_salary)
```

- CA has significantly higher avg salary than other states
- IL, DC and MA also have considerably higher avg salary than other states

### c. Number of jobs in different states

```python
job_openings = df['Job Location'].value_counts()[0:10]
job_openings = job_openings.reset_index().rename(columns={'index':'State', "Job Location":"Number of Job Openings"})
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.barplot(x='State', y='Number of Job Openings', data=job_openings)
```

- CA has the most job openings because it has the biggest tech hub in the nation.
- Suprisingly, MA has more job openings than NY even though having less Fourtone 500 headquarters in the state.
- Even though IL pays relatively higher salary than most of the states, It is not in the four states having the most job openings.

### d. Industries having the highest job openings

```python
industry = df['Industry'].value_counts()[0:15]
industry = industry.reset_index().rename(columns={'index':'Industry', "Industry":"Number of Job Openings"})
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.barplot(y='Industry', x='Number of Job Openings', data=industry)
```

Interestingly, Biotech & Pharmaceutical industry has the highest job openings, even higher than technology.

### e. How education level affects salary? 

```python
education = df['Degree'].str.replace('na', 'Undergraduate', regex=False)
education = education.str.replace('M', 'Masters', regex=False)
education = education.str.replace('P', 'Phd', regex=False)
```

```python
education = education.value_counts()
education = education.reset_index().rename(columns={'index':'Education', "Degree":"Count"})
education
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.barplot(x='Education', y='Count', data=education)
```

About half of the data science jobs mentions graduate degree(masters or phd) in their description.

### f. What tools are the most important in different job titles?


```python
tools = ['Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit',
       'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an']
```

```python
data_tools = df.groupby("job_title_sim")['Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit',
       'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an'].sum()
data_tools1 = data_tools.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
tools = ['Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit',
       'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an']
for i in tools:
    ax = sns.barplot(x=i, y='job_title_sim', data=data_tools1)
    plt.figure()
```

The graphs show that python, excel, sql, tableau, aws and hadoop are the most mentioned tools in job postings.
